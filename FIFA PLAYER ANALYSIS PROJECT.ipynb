### Missing Values
When exploring data, you’ll most likely encounter missing or null values, which are essentially placeholders for non-existent values. Most commonly you'll see Python's None or NumPy's np.nan, each of which are handled differently in some situations.

#### There are mainly three types of missing values.
#### 1.MCAR(Missing completely at random): These values do not depend on any other features.
#### 2.MAR(Missing at random): These values may be dependent on some other features.
#### 3.MNAR(Missing not at random): These missing values have some reason for why they are missing.
Let’s see which columns have missing values in the dataset.

```python
# Checking the missing values
df.isnull().sum()
```

#### As we can see there are columns contain missing values. Let’s see how to handle the missing values. We can handle missing values by dropping the missing records or by imputing the values.

But before doing that, let us check whether there are any rows with non_null values(in atleast one column) or not.

```python
# len(df) == df.shape[0]
len(df) - len(df[df.isnull().any(axis = 1)])
```

#### There are no rows with non_null values(in atleast one column). So we cannot drop any row.
#### So, let us impute the values.

### Imputation
#### Imputation is a conventional feature engineering technique used to keep valuable data that have null values.
There may be instances where dropping every row with a null value removes too big a chunk from your dataset, so instead we can impute that null with another value, usually the mean, mode or the median of that column.

```python
# First of all, let us see the no. of columns containing null values.
cols = df[df.columns[df.isnull().any()]].columns
len(cols)
```

#### Now let us extract the names of cateogerical columns and numerical columns into seperate variables.
Cateogerical columns are nothing but the columns containing the values other than numerical values i.e containing data of type object(strings are one of the examples)

#### But before doing that, first we have to handle the missing values in our 'Output column'/'Target column'.

Iam considering 'Release_Clause' as the target column.

Handling the missing values in the 'Release_Clause' column. Since, our target column is 'Release_Clause' Column, if we impute the values to this column it’ll affect our analysis. So, it is better to drop the missing values from 'Release_Clause' Column.

```python
#drop the records with 'Release_Clause' missing in data.
df = df[~df['Release_Clause'].isnull()]
```

```python
df.shape
```

```python
df.info()
```

If we observe, 'Loaned_From' column has 0 non-null values, So, let us drop that column from our data frame.

```python
df = df.drop('Loaned_From', axis = 1)
df.info()
```

```python
# Checking the missing values
df.isnull().sum()
```

Now let us extract the names of cateogerical columns and numerical columns into seperate variables.

```python
numeric_cols = df._get_numeric_data().columns
numeric_cols
```

```python
cat_cols = df.drop(numeric_cols, axis = 1).columns
cat_cols
```

```python
# As 'Release_Clause' is target column, no need to extract that into cat_cols.
cat_cols = cat_cols.drop('Release_Clause')
cat_cols
```

#### Now fill/impute the null values in numerical columns with mean of the respective column.

```python
# Let us store the mean of each numeric column into a variable.
numeric_mean = df[numeric_cols].mean()
numeric_mean
```

```python
df[numeric_cols] = df[numeric_cols].fillna(numeric_mean)
df[numeric_cols].isnull().sum()
```

#### Now fill/impute the null values in cateogerical columns with mode of the respective column.

```python
# Let us store the mode of each cateogerical column into a variable.
cat_mode = {}
for col in cat_cols:
    cat_mode[col] = df[col].mode()[0]
    
cat_mode = pd.Series(cat_mode)
cat_mode
```

```python
df[cat_cols] = df[cat_cols].fillna(cat_mode)
df[cat_cols].isnull().sum()
```

```python
df.isnull().sum().sum()
```

Finally, we have handled the missing values by either removing them or imputing with some other values like mean, mode.

### Handling Outliers
We have seen how to fix missing values, now let’s see how to handle outliers in the dataset.

#### Outliers are the values that are far beyond the next nearest data points.

There are two types of outliers:

#### 1)Univariate outliers: Univariate outliers are the data points whose values lie beyond the range of expected values based on one variable.
#### 2)Multivariate outliers: While plotting data, some values of one variable may not lie beyond the expected range, but when you plot the data with some other variable, these values may lie far from the expected value.

So, after understanding the causes of these outliers, we can handle them by dropping those records or imputing with the values or leaving them as is, if it makes more sense.

Here we convert the height strings into centimeter values. We do this by multiplying the first element of the string, for example ‘5’7’, by 12.0 (inches) and we add the third element of the string ‘7’ (which is in inches) and multiply the final result by 2.54 (centimeters):

### c) Categorical — Categorical Analysis

Since our target variable/column is the 'Release_Clause', we’ll see how the different categories like 'Preferred_Foot', 'Work_Rate','Position', etc., are associated with the 'Release_Clause' column.

```python
def to_numeric_fun(x):
    size = len(x)
    y = float(x[1:size-1])
    if(x[size-1] == 'M'):
        y = y*1000000
    else:
        y = y*1000
    return y

target_numeric_values = df['Release_Clause'].apply(to_numeric_fun)
target_mean = target_numeric_values.mean()
print(target_mean)
```

```python
# So, let us create another column by 
# converting the values in 'Release_Clause' into 1 or 0.
# This can be done by taking the mean of the column and 
# if the value is > mean, then it'll be converted to 1 or else 0

def convert_fun(x):
    mean = 4585060.986600974
    if(x >= mean):
        return 1
    else:
        return 0

# Let us name the new column as 'Target'
df['Target'] = target_numeric_values.apply(convert_fun)
```

```python
#plot the bar graph of 'Preferred_Foot' with average value of 'Target'
plt.figure(figsize = (5,3))
df.groupby('Preferred_Foot')['Target'].mean().plot.bar()
plt.show()
```

By the above graph, we can infer that the earning is more for members with 'Preferred_foot" as 'Left' in the data set. Similarly we can plot for 'Work_Rate','Position',etc...,.

```python
#plot the bar graph of 'Work_Rate' with average value of 'Target'
plt.figure(figsize = (6,4))
df.groupby('Work_Rate')['Target'].mean().plot.bar()
plt.show()
```

```python
#plot the bar graph of 'Position' with average value of 'Target'
plt.figure(figsize = (8,5))
df.groupby('Position')['Target'].mean().plot.bar()
plt.show()
```

### 5. Multivariate Analysis
If we analyze data by taking more than two variables/columns into consideration from a dataset, it is known as Multivariate Analysis.

Let’s see how 'Preferred_Foot', 'Work_Rate', and 'Target' vary with each other.

First, we’ll create a pivot table with the three columns and after that, we’ll create a heatmap.

```python
result = pd.pivot_table(data=df, index='Preferred_Foot',
                        columns='Work_Rate',values='Target')
result
```

```python
#create heat map of 'Preferred_Foot' vs 'Work_Rate' vs 'Target'
sns.set(rc = {'figure.figsize':(8,4)}, font_scale = 1.2)
sns.heatmap(result, annot=True);
```

Based on the Heatmap we can infer that the players with 'Work_Rate':'Low/Low' and 'Preferred_Foot':'Left' are less likely to earn more and 'Work_Rate':'High/High' and 'Preferred_Foot':'Left'/'Right' are most likely to earn more.
